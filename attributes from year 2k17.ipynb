```python
dfMerged['genres'] = dfMerged["genres"].map(lambda x: ast.literal_eval(x))
dfMerged["crew"] = dfMerged["crew"].map(lambda x: ast.literal_eval(x))
dfMerged["cast"] = dfMerged["cast"].map(lambda x: ast.literal_eval(x))
```

```python
def genres(x):
    gen = []
    st = " "
    for i in x:
        if i.get('name') == 'Science Fiction':
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen == []:
        return np.NaN
    else:
        return (st.join(gen))
```

```python
def actors(x, num):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == []:
        return np.NaN
    if num == 1:
        return (casts[0])
    elif num ==2:
        if len(casts)<=1:
            return np.NaN
        return (casts[1])
    elif num ==3:
        if  len(casts)<=2:
            return np.NaN
        else:
            return (casts[2])
```

```python
dfMerged['genres'] = dfMerged["genres"].map(lambda x: genres(x))
dfMerged['actor_1_name'] = dfMerged['cast'].map(lambda x: actors(x,1))
dfMerged['actor_2_name'] = dfMerged['cast'].map(lambda x: actors(x,2))
dfMerged['actor_3_name'] = dfMerged['cast'].map(lambda x: actors(x,3))
```

```python
dfMerged.drop("cast",axis=1,inplace=True)
```

```python
dfMerged.head()
```

```python
def directors(x):
    dt = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    else:
        return (st.join(dt))
```

```python
dfMerged['director_name'] = dfMerged['crew'].map(lambda x: directors(x))
dfMerged.drop("crew", axis=1,inplace=True)
```
